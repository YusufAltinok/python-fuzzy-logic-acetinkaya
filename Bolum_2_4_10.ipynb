```python
# Veriler dosyasının yolu
veriler_path = '/content/drive/MyDrive/Colab Notebooks/veriler.csv'
```

```python
veri = pd.read_csv(veriler_path)
print(veri.columns)

boy = veri['boy']
print(boy)

boy = veri[['boy']]
print(boy)

boykilo = veri[['boy', 'kilo']]
boykilo.columns = ['Boy D. (cm)', 'Kilo D. (kg)']
print(boykilo)

plt.plot(boykilo)
plt.legend(('Boy D. (cm)', 'Kilo D. (kg)'), loc='upper left') 
plt.title('Veri Tablosu - Boy & Kilo')
plt.ylabel('Y Ekseni - Veri Ölçüm Değeri')
plt.xlabel('X Ekseni - Dizi İçerisinde Kişi Sıra Bilgisi')
plt.figure(figsize=(10,5))
plt.show()
```
